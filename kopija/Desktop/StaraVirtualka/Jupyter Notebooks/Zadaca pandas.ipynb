```python
import pandas as pd
import numpy as np
```

```python
'Create a new column called "diff". This column returns 1 if column num > column num2 and 0 if it does not. What is the sum of column diff'
```

```python
df=pd.read_csv('Pandas Bootcamp.csv')
df
```

```python
df['diff']=np.where(df['num']>df['num2'],1,0)
```

```python
df['diff'].sum()
```

```python
'Find the least popular string in column str2. Filter by str2 where the string equals the least popular string. Find the sum of column num.'
```

```python
df2=pd.read_csv('Pandas Bootcamp2.csv')
df2
```

```python
df2['num'][df2['str2']==df2['str2'].value_counts().idxmin()].sum()
```

```python
'Create a new column called "Letter". This column is the sum of the lengths of columns str and str2. Then divide the column num by Letter. Pick the second smallest number.'
```

```python
df5=pd.read_csv('Pandas Bootcamp5.csv')
df5
```

```python
df5['Letter']=df5['str'].str.len()+df5['str2'].str.len()
```

```python
(df5['num']/df5['Letter']).nsmallest(2).iloc[-1]
```
