Kako bih nacrtao histograme podataka zajedno s grafom normalne razdiobe definiram funkciju crtaj_histogram.

```python
def crtaj_histogram(data, stupci={}):
    '''Funkcija prima DataFrame i crta histograme frekvencija svih stupaca zajedno sa funkcijom gustoce normalne razdiobe.
    Dodatni (opcionalni) parametar stupci je dictionary kojim se mogu zadavati imena histograma različita od imena stupaca u tablici.
    Ako je parametar stupci prazan rjecnik, onda su imena histograma imena stupaca u tablici.'''
    
    test = lambda br_stupaca : int(br_stupaca/2)+1 if br_stupaca % 2 else int(br_stupaca/2)
    m = test(len(data.columns)) #Broj redaka u subplots
    
    hist, ax = plt.subplots(m, 2, figsize=(15,15))
    hist.tight_layout(pad=4.0)
    j=0
    for row in ax:
        for axis in row:
            if (j<len(data.columns)):
                axis.hist(data.iloc[:,j], density=True, ec = 'black')
                
                mu, sigma = data.iloc[:,j].mean(), data.iloc[:,j].std()
                x = np.linspace(mu - 3*sigma, mu+3*sigma) #znamo da je u normalnoj distribuciji preko 99% podataka na udaljenosti manjoj od 3*std
                y = scs.norm.pdf(x, mu, sigma)
                axis.plot(x, y, color='red', lw=1.4)
                
                if not bool(stupci): #ako je dict prazan, ispisujem imena stupaca
                    axis.set_title(data.columns[j])
                else: #inace koristi dict   
                    axis.set_title(stupci[data.columns[j]])
                axis.set_ylabel('Frequency')
                axis.set_xlabel('')
            else:
                axis.remove()
            j+=1
```

```python
crtaj_histogram(data, stupci)
```


## Analiza normalnosti podataka

```python
def crtaj_norm_vjer_graf(data, stupci={}):
    '''Funkcija prima DataFrame i crta normalne vjerojatnosne grafove
    Dodatni (opcionalni) parametar stupci je dictionary kojim se mogu zadavati imena histograma različita od imena stupaca u tablici.
    Ako je parametar stupci prazan rjecnik, onda su imena histograma imena stupaca u tablici.'''
    test = lambda br_stupaca : int(br_stupaca/2)+1 if br_stupaca % 2 else int(br_stupaca/2)
    m = test(len(data.columns)) #Broj redaka u subplots
    
    fig, ax = plt.subplots(m, 2, figsize=(15,15))
    fig.tight_layout(pad=5.0)
    j=0
    for row in ax:
        for axis in row:
            if (j<len(data.columns)):
                stats.probplot(data.iloc[:,j], plot=axis, rvalue=True, dist='norm')
                if not bool(stupci):
                    axis.set_title(data.columns[j])
                else:   
                    axis.set_title(stupci[data.columns[j]])
            else:
                axis.remove()
            j+=1
```

```python
crtaj_norm_vjer_graf(data,stupci)
```

Uzorci koji dolaze iz normalne distribucije moraju aproksimirati nacrtani pravac. Važno je uočiti odstupaju li repovi od kretanja pravca i postoji li neki uzorak u rasporedu točaka.

```python
def boja(pv):
    boja = 'red' if pv < 0.05 else 'black'
    return 'color: %s' % boja

normal_test = pd.DataFrame([[lilliefors(data[stupac], dist='norm', pvalmethod='table')[1] for stupac in data.columns],
                            [stats.normaltest(data[stupac])[1] for stupac in data.columns],
                            [stats.shapiro(data[stupac])[1] for stupac in data.columns]],
                            columns=data.columns,
                            index=['Lilliefors KS-test', 'normaltest', 'Shapiro'])

normal_test.style.applymap(boja).set_precision(4)
```

P-vrijednosti testova normalnosti su dane u gornjoj tablici. Značajne vrijednosti na razini značajnosti $\alpha = 0.5$ označene su crvenom bojom.

Na temelju grafičkih testova i gornja tri testa normalnosti donosim zaključke o normalnosti. ### Koeficijenti korelacije

Sada računam koeficijente korelacije između parova podataka danih u tablici.
Prvo računam Pearsonov koeficijent korelacije i testiram hipotezu $H_0: \rho = 0$ nasuprot hipotezi $H_1: \rho \neq 0$.
Za testiranje Pearsonovog koeficijenta nužna je pretpostavka o normalnosti podataka. Zato za provjeru računam i Spearsonov koeficijent jer analogan test [ne zahtjeva](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.spearmanr.html) pretpostavku o normalnosti. Pearsonov koeficijent korelacije pokazuje stupanj linearnog odnosa između podataka, a Spearmanov koeficijent stupanje monotone povezanosti podataka ([izvor](https://support.minitab.com/en-us/minitab-express/1/help-and-how-to/modeling-statistics/regression/supporting-topics/basics/a-comparison-of-the-pearson-and-spearman-correlation-methods/)).


```python
def korelacija(data, metoda):
    '''Funkcija računa matricu koeficijenata korelacija nad stupcima DataFrame-a.
    Koeficijenti koji su značajni na razini značajnosti 0.05 označeni su s ***. Argumenti funkcije su DataFrame podataka i
    string koji označava metodu koristim'''
    if metoda not in {'pearson','spearman'}:
        raise ValueError("argument mora biti 'pearson' ili 'spearman'")
    cor = data.corr(method=metoda)
    for i in data.columns:
        for j in data.columns:
            if metoda == 'pearson':
                ro, pvalue = stats.pearsonr(data[i], data[j])
            else:
                ro, pvalue = stats.spearmanr(data[i], data[j])
            if i == j:
                cor.loc[i,j] = ''
            elif pvalue < 0.05:
                cor.loc[i,j] = '***%.4f' % (cor.loc[i,j])
            else:
                cor.loc[i,j] = '%.4f' % (cor.loc[i,j])
    return cor
```

```python
korelacija(data, 'pearson')
```

```python
korelacija(data, 'spearman')
```

Oznaka *** uz koeficijent korelacije u tablicama znači da je taj rezultat statistički značajan na razini značajnosti 0.05, odnosno koeficijent korelacije za te dvije varijable je različit od 0.

### Linearna regresija

Radim linearnu regresiju odnosa BDP-a per capita i duljine trajanja života. Iz dijagrama raspršenja vidim da bi bilo dobro logaritmirati podatke o BDP-u per capita kako bih dobio linearan odnos.

```python
import statsmodels.api as sm
x = np.log(data['GDP per capita']).values.reshape(-1,1)
y = data['Life expectancy'].values.reshape(-1, 1)
x = sm.add_constant(x)

model = sm.OLS(y, x)
res=model.fit()
res.summary()
```

Iz tablice rezultata vidim da je F-statistika jednaka 311.3, a p-vrijednost jako mala pa je model značajan. Koeficijent determinacije iznosi 74% što znači da je 74% varijabilnosti objašnjeno ovim modelom. Koeficijenti u modelu su značajni.

```python
'y = {} + {}*x'.format(res.params[0].round(4), res.params[1].round(4))
```

Za linearni regresijski model važno je provjeriti pretpostavku da su reziduali normalno distribuirani, te nacrtati model kako bi provjerili je li model dobar. Naime, može se dogoditi da su gornji pokazatelji dobri, ali da model nije dobar.

```python
fig, ax=plt.subplots(1,2, figsize=(15,5))
sns.residplot(x[:,1], y, ax=ax[0]);
ax[0].set_title("Graf reziduala")
ax[0].set_xlabel("log(GDP per capita)")
ax[0].set_ylabel("Reziduali")

sns.regplot(x[:,1], y,ax=ax[1]);
ax[1].set_title("Linearni regresijiski model")
ax[1].set_xlabel("log(GDP per capita)")
ax[1].set_ylabel("Duljina života");
```

Iz grafa rezuduala vidim da su reziduali uglavnom nasumično raspoređeni. Međutim, negativni reziduali imaju veći raspon od pozitivnih. Ovakav graf može ukazivati na to da reziduali nisu normalno distribuirani oko 0 s istim varijancama. Također, za veće vrijednosti BDP-a per capita, reziduali su negativni. Ovo nije toliko uočljivo na ovom prikazu, ali je na grafu s originalnim podacima. Sljedeći graf se može zoomirati i pogledati imena država na koje se dana točkica odnosi. ### ANOVA

[ANOVA](https://pythonfordatascience.org/anova-python/) je test jednakosti očekivanja za različite populacije. Dakle, nezavisna varijabla je diskretna, a zavisna normalno distribuirana. Želim promatrati postotak potrošnje BDP-a na obrazovanje i broj gradskog stanovništva u odnosu na bogatstvo države. Podijelit ću države na četiri grupe po visini BDP-a per capita tako da ih je u svakoj grupi jednako mnogo. Za to su mi potrebni kvantili podataka 'GDP per capita'. Dobivam ordinalnu varijablu 'Grupe_GDP'.

```python
deskriptiva = data['GDP per capita'].describe()
```

```python
def f(data):
    if data['GDP per capita'] < deskriptiva['25%']:
        val = 0
    elif deskriptiva['25%'] <= data['GDP per capita'] < deskriptiva['50%']:
        val = 1
    elif deskriptiva['50%'] <= data['GDP per capita'] < deskriptiva['75%']:
        val = 2
    else: val = 3
    return val

data['Grupe_GDP'] = data.apply(f, axis=1)
```

```python
data.rename(columns={'Exp on edu (% of GDP)': 'ExpGDP',
                    'Urban (%)': 'Urban'},
                    inplace=True)
```

Prvo promatram postotak potrošnje BDP-a na obrazovanje u odnosu na grupe država prema visini BDP-a. Prikažimo podatke grafički.

```python
sns.boxplot(data['Grupe_GDP'], data['ExpGDP']);
```

Iz dijagrama pravokutnika vidim da se zavisna varijabla blago raste s povećanjem BDP-a. ANOVA će dati odgovor je li taj rast značajan.

Trebam provjeriti pretpostavke za ANOVU. U analizi normalnosti nisam odbacio pretpostavku da je postotak potrošnje BDP-a na obrazovanje normalno distribuiran. Trebam još provjeriti jednakost varijanci. Za to koristim Levenov test za homogenost varijanci. Nulta hipoteza je da su varijance homogene.

```python
param = (data['ExpGDP'][data['Grupe_GDP']==i] for i in data['Grupe_GDP'].drop_duplicates().values)
levene_res = [list(stats.levene(*param))]
pd.DataFrame(levene_res, columns=['statistika', 'p-vrijednost'], index=['Levene test'])
```

```python
a = data.groupby('Grupe_GDP')['ExpGDP'].var()
print(round(a.max()/a.min(), 3))
```

U Levenovom testu odbacujem hipotezu da su varijance homogene, međutim ANOVA se može koristiti i kad razlike u varijancama nisu prevelike kao što je detaljno objašnjeno [ovdje](https://link.springer.com/article/10.3758/s13428-017-0918-2). U našem slučaju omjer najveće i najmanje varijance je 2.661 pa provodim ANOVU.

```python
from statsmodels.formula.api import ols
rez = ols('ExpGDP ~ C(Grupe_GDP)', data=data).fit()
anova_tablica = sm.stats.anova_lm(rez)
anova_tablica
```

ANOVA je ekvivalentna linearnom regresijskom modelu pa sam ju izvršio pozivom linearnog regresijskog modela s dummy varijablama te ispisom standardne tablice za ANOVU. Zaključujem da se postotak potrošnje BDP-a na obrazovanje značajno razlikuje za države podijeljene prema visini BDP-a per capita. Kako bih detaljnije provjerio koje grupe rade razliku provodim [Tukeyjev post-hoc test](https://www.statisticshowto.com/tukey-test-honest-significant-difference/).

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.multicomp import MultiComparison

mc = MultiComparison(data['ExpGDP'], data['Grupe_GDP'])
tukey_res = mc.tukeyhsd()
print(tukey_res)
```

Zaključujem da samo prva i zadnja grupa, odnosno države s najmanjim i najvećim BDP-om per capita, značajno razlikuju u postotku BDP-a koji se troši na obrazovanje.

Napravimo sada analizu postotka gradskog stanovništva u odnosu na grupe država.

```python
sns.boxplot(data['Grupe_GDP'], data['Urban']);
```

```python
param = (data['Urban'][data['Grupe_GDP']==i] for i in data['Grupe_GDP'].drop_duplicates().values)
levene_res = [list(stats.levene(*param))]
pd.DataFrame(levene_res, columns=['statistika', 'p-vrijednost'], index=['Levene test'])
```

U Lillieforovoj inačici KS testa za normalnost nisam odbacio pretpostavku da podaci za postotak gradskog stanovništva dolaze iz normalno distribuirane populacije. Levenovim testom ne odbacujem pretpostavku da su varijance homogene. Dakle, zadovoljene su pretpostavke ANOVE.

```python
rez = ols('Urban ~ C(Grupe_GDP)', data=data).fit()
anova_tablica = sm.stats.anova_lm(rez)
anova_tablica
```

P-vrijednost je jako mala, pa na standardnim razinama značajnosti odbacujem pretpostavku da su očekivanja jednaka.

```python
mc = MultiComparison(data['Urban'], data['Grupe_GDP'])
tukey_res = mc.tukeyhsd()
print(tukey_res)
```

Prema Tukeyjevom post hoc testu razlike među svim grupama su značajne. Kako je Grupe_GDP ordinalna varijabla, tj. postoji uređaj među njima, napravit ću i linearni regresijski model na istim podacima kako bih otkrio je li veza između postotka gradskog stanovništva i grupa država linearna.

```python
X = data['Grupe_GDP']
Y = data['Urban']
X=sm.add_constant(X)
model= sm.OLS(Y,X).fit()
res = model.summary()
res
```

```python
'y = {} + {}*x'.format(model.params[0].round(4), model.params[1].round(4))
```

P-vrijednost u testu značajnosti modela (Prob (F-statistic)) je mala pa je test značajan. Koeficijent determinacije iznosi 64% što znači da je 64% varijabilosti opisano danim modelom. Iz procijenjenih parametara zaključujem da se u prosjeku udio gradskog stanovništva poveća za 16.8% prelaskom u sljedeću grupu po visini BDP-a per capita pri čemu je u grupi najsiromašnijih država postotak gradskog stanovništva procijenjen sa 32.4%. Ovo je prikazano u sljedećem grafu.

Postotak potrošnje na obrazovanje koji se odnosi na plaće zaposlenika, kratkoročna ulaganja, odnosno dugoročna ulaganja u obrazovanje nisu korelirana s pokazateljima razvijenosti zemalja. Postotak gradskog stanovništva, prosječna duljina života, BDP per capita su očekivano jako pozitivno korelirani.

Najbogatije države ulažu značajno viši postotak BDP-a u obrazovanje od najsiromašnijih a razlika u ulaganju sličnijih država prema bogatstvu nije značajna.
