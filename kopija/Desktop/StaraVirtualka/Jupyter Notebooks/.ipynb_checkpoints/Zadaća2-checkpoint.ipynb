```python
def pomocna(x,tocke,radijusi):
    for (tocka,radijus) in zip(tocke,radijusi):#paralelno ide kroz ovo
        if((x.real-tocka.real)**2 +(x.imag-tocka.imag)**2 <=radijus**2):
            return True
        else:
            continue
    return False
```

```python
def is_covered(*args):
    x=args[0][0]
    y=args[0][1]
    l=np.random.uniform(0,x,1000000)+np.random.uniform(0,y,1000000)*1j
    argsi=np.array(args[1])
    tocke=argsi[:,0]+argsi[:,1]*1j
    radijusi=argsi[:,2]
    polje=np.array([pomocna(tocka,tocke,radijusi) for tocka in l])
    if len(polje[polje == False])>0:
        return False
    else:
        return True
```
