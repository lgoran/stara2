# Numeričko rješavanje inicijalnog problema za ODJ

U mnogim primjenama u prirodnim i inženjerskim znanostima se vremenska promjena stanja promatranog sustava (mehanički sustav, kemijska reakcija) opisuje diferencijalnim jednadžbama. One matematički opisuju zakone (npr. zakoni mehanike) koji se prikazuju na sljedeći način:

$$y'(t)=f(t,y(t)),\quad t\geq t_0 \\
y(t_0)=y_0 \tag{1}$$ 

Pri tome je $f : I × \mathbb{R}^d\longrightarrow \mathbb{R}^d$ neprekidna i $I \subseteq \mathbb{R}$ otvoren interval, $t_0 ∈ I, d \geq 1$ i
$y_0 \in \mathbb{R}^d$

## Jednokoračne metode

Riješiti problem (1) numerički znači izračunati aproksimacije vrijednosti $y(t_i)$
u konačno točaka $t_1,...,t_n$ u zadanom intervalu $[t_0, T]$.

Ključni element numeričkog rješavanja inicijalnog problema (1) je kako iskoristiti diferencijalnu jednadžbu i od poznate vrijednosti $y(t_0)$ dobiti što bolju aproksimaciju za $y(t_1)$ i tako dalje, $t_0 \to t_1 \to\ldots\to t_i \to t_{i+1} \to \ldots \to t_n$sve do zadnje točke. Kao što vidimo, za dobivanje $y(t_{i+1})$ koristimo samo jednu već poznatu vrijednost i to $y(t_i)$, tj. samo jedan korak, odakle i ime jednokoračne metode.

### Eulerova Metoda

Iz matematičke analize znamo da vrijedi sljedeća relacija:

$$
y(t_{i+1}) - y(t_i) = \int_{t_i}^{t_{i+1}} y'(\tau) d\tau = \int_{t_i}^{t_{i+1}} f(\tau,y(\tau)) d\tau
$$

Iz čega slijedi:
$$
y(t_{i+1}) = y(t_i) + \int_{t_i}^{t_{i+1}} f(\tau,y(\tau)) d\tau
$$

Aproksimiranjem integrala $\int_{t_i}^{t_{i+1}} f(\tau,y(\tau)) d\tau \approx (t_{i+1} - t_i)f(t_i, y(t_i) ) $ dobivamo $y(t_{i+1}) \approx y(t_i) + (t_{i+1} - t_i)f(t_i, y(t_i) )$ , što je upravo **glavna ideja** Eulerove metode

Slijedi nam implementacija metode. Prije svega uvezimo potrebne pakete:

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.max_open_warning'] = 50
%matplotlib agg
import time
import numpy as np
import pandas as pd
import math
from pprint import pprint
```

Funkcija **feval**(**F**inite Element **Eval**uator) služi za, kako samo ime kaže, evaluaciju funkcije na konačno mnogo točaka.

```python
def feval(funcName, *args):
    return eval(funcName)(*args)
```

Eulerova metoda prima funkciju `f`, početni uvijet `y0`, interval na kojem radimo `[a,b]` i fiksni korak `h`.
Funkcija vraća ekvidistantnu mrežu točaka `x` te rješenje inicijalnog problema u tim točkama `y`.

```python
def EulerovaMetoda(f, y0, a, b, h):
    broj_podintervala = int((b - a)/h)    
    xradni = a
    yradni = y0    
    x = np.array([xradni])
    y = np.array([yradni[0]])
    
    for i in range(broj_podintervala):
        fja = feval(f, xradni, yradni)    
        yradni += h*np.array(fja)         
        xradni += h
        x = np.append(x,xradni)   
        y = np.append(y,yradni)           
    return [x, y]
```

Kako bismo mogli testirati Eulerovu metodu, trebamo uzeti neku jednadžbu za koju znamo egzaktno rješenje. Kada ga ne bismo znali, ne bismo mogli adekvatno ocijeniti točnost metode. Neka je zato inicijalni problem:
$$
y'(x) = -y(x) - 5e^{-x}\sin(5x),\quad x\in [0,3]\\
y(0) = 1,$$

čije je egzaktno rješenje dano sa $y(x) = e^{-x}\cos(5x)$.
Radi jednostavnijeg uspoređivanja sve metode ćemo promatrati upravo na gornjem problemu.

```python
def testFunkcija(x, y):
    dy = [0] * len(y)
    dy[0] = -y[0] - 5*math.exp(-x)*math.sin(5*x)
    return dy
```

Želimo obraditi nekoliko metoda i nakon toga ih međusobno usporediti. Zbog toga ih testiramo na jednak način. Prvo ih izvršimo, mjerući vrijeme izvršavanja, zatim usporedimo sa egzaktnim rješenjem i utvrdimo iznos maksimalne pogreške.

Svakako ćemo htjeti i vizualno usporediti egzaktno rješenje s aproksimacijama.

```python
def testProtokol(metoda,ime_metode):
    h = 0.1
    a = 0
    b = 3
    y0 = np.array([1.0])
    start_time=time.time()
    [x, y] = metoda('testFunkcija', y0, a, b, h)
    vrijeme_izvršavanja=time.time()-start_time
    n = int((b - a) / h)
    t = np.array([a+i*h for i in range(n+1)])
    rješenje = np.empty(0)
    greška=np.empty(0)
    for i in range(n+1):
        pom = math.exp(-t[i])*math.cos(5*t[i])
        rješenje = np.append(rješenje,pom)
        greška = np.append(greška,abs(pom-y[i]))
    pprint("Maksimalna greška: " + str(max(greška)))
    pprint("Vrijeme izvršavanja: " + str(vrijeme_izvršavanja*1000)+ "ms")
    fig,ax=plt.subplots()
    ax.plot(x, y, 'ro--')
    ax.plot(t, rješenje, 'b*-')
    ax.set_xlim(a, b)
    ax.legend([ime_metode,"Egzaktno rješenje"])
    ax.set_xlabel('x', fontsize=20)
    ax.set_ylabel('y', fontsize=20)
    return [ime_metode,vrijeme_izvršavanja*1000,max(greška),fig]
```

```python
#testiranje na primjeru
[ime_metode,vrijeme_izvršavanja,greška,fig]=testProtokol(EulerovaMetoda,"Eulerova Metoda")
podaci=np.array([[ime_metode,vrijeme_izvršavanja,greška]])
fig
```

Vidjeli smo da Eulerova metoda nije baš najsretnije rješenje s obzirom na veličinu greške.
**Modificirana Eulerova metoda** je zato ipak dosta preciznija. Koristi se sličan princip kao i u originalnoj metodi, ali bitno poboljšanje dolazi u obliku traženja aproksimacije na polovištu referentnog intervala. Zbog toga se još zna nazivati i *Midpoint method*.

```python
def ModificiranaEulerovaMetoda(f, y0, a, b, h):
    broj_podintervala = int((b - a)/h)    
    xradni = a
    yradni = y0    
    x = np.array([xradni])
    y = np.array([yradni[0]])
    
    for i in range(broj_podintervala):
        fja = feval(f, xradni, yradni)
        fi = feval(f, xradni+h/2, yradni+np.array(fja)*h/2)
        yradni += h*np.array(fi)         
        xradni += h
        x = np.append(x,xradni)  
        y = np.append(y,yradni)
    return [x, y]
```

```python
#testiranje na primjeru
[ime_metode,vrijeme_izvršavanja,greška,fig]=testProtokol(ModificiranaEulerovaMetoda,"Modificirana Eulerova Metoda")
podaci = np.append(podaci,[ime_metode,vrijeme_izvršavanja,greška])
fig
```

Maksimalna greška je sada znatno smanjena. Možemo zaključiti da je modificirana Eulerova metoda puno bolje rješenje od originala i za sada najbolje što imamo, ali ne i zadugo...

### Runge–Kuttine metode

Željeli bismo poboljšati jednostavne jednokoračne metode. Polazimo od osnovne relacije
$$
y(t_{i+1}) = y(t_i) + \int_{t_i}^{t_{i+1}} f(\tau,y(\tau)) d\tau
$$

i u njoj pokušajmo numeričku vrijednost integrala aproksimirati koristeći dodatne čvorove u $[t_k, t_{k+1}]$ i dodatne slobodne parametre koje ćemo naknadno fino naštimati. Konkretno, uzmimo na primjer tri čvora $\varepsilon_1, \varepsilon_2, \varepsilon_3$, tri dodatna parametra $c_1, c_2, c_3$ i potražimo formulu oblika

$$y_{k+1} = y_k + h[c_1f(\varepsilon_1,y(\varepsilon_1)) + c_2f(\varepsilon_2,y(\varepsilon_2)) +c_3f(\varepsilon_3,y(\varepsilon_3))]$$

pri čemu ćemo slobodne parametre iskoristiti da dobijemo što je moguće manju lokalnu pogrešku.

Jedna od najjednostavnijih Runge-Kuttinih metoda je Heunova. Ona je reda 2 što znači da evaluira funkciju f 2 puta. Prva točka je početna točka referentnog intervala dok je druga krajnja uz odgovarajuće koeficijente $c_1 =\frac{1}{2}$ i $c_2 = \frac{1}{2}$, respektivno. Zbog svoje jednostavnosti nam daje dobar uvid kako se Runge–Kuttine metode općenito implementiraju.

```python
def HeunovaMetoda(f, y0, a, b, h):
    broj_podintervala = int((b - a)/h)    
    xradni = a
    yradni = y0    
    x = np.array([xradni])
    y = np.array([yradni[0]])

    for i in range(broj_podintervala):
        k1 = np.array(feval(f, xradni, yradni))
        k2 = np.array(feval(f, xradni+h, yradni + np.array(k1) * h))
        yradni += h/2*(k1+k2)
        xradni += h
        x = np.append(x,xradni)
        y = np.append(y,yradni)
    return [x, y]
```

```python
#testiranje na primjeru
[ime_metode,vrijeme_izvršavanja,greška,fig]=testProtokol(HeunovaMetoda,"Heunova Metoda")
podaci = np.append(podaci,[ime_metode,vrijeme_izvršavanja,greška])
fig
```

Možda i najpopularnija metoda je svakako Runge-Kutta reda 4 (**rk4**). Točke u kojima se evaluira funkcija su početna točka referentnog intervala, zatim dva puta središnja točka , te na kraju krajnja točka intervala sa koeficijentima $c_1 = \frac{1}{6}, c_2 = \frac{1}{3},c_3 = \frac{1}{3},c_4 = \frac{1}{6}$, respektivno.
Metoda se implementira vrlo slično kao i Heunova, odnosno slično kao i sve Runga-Kuttine metode.

```python
def rk4(f, y0, a, b, h):
    broj_podintervala = int((b - a)/h)    
    xradni = a
    yradni = y0    
    x = np.array([xradni])
    y = np.array([yradni[0]])

    for i in range(broj_podintervala):
        k1 = np.array(feval(f, xradni, yradni))
        k2 = np.array(feval(f, xradni+h/2, yradni + np.array(k1) * h/2))
        k3 = np.array(feval(f, xradni+h/2, yradni + np.array(k2) * h/2))
        k4 = np.array(feval(f, xradni+h, yradni + np.array(k3) * h))
        
        yradni += h/6*(k1+2*k2+2*k3+k4)
        xradni += h
        x = np.append(x,xradni)
        y = np.append(y,yradni)
    return [x, y]
```

```python
#testiranje na primjeru
[ime_metode,vrijeme_izvršavanja,greška,fig]=testProtokol(rk4,"rk4 Metoda")
podaci = np.append(podaci,[ime_metode,vrijeme_izvršavanja,greška])
fig
```

Na maksimalnoj greški vidimo zašto je ova metoda jedna od popularnijih. Greška je drastično smanjena naprema svim prošlim metodama uz ne preveliki utrošak vremena.

## Višekoračne Metode

Kada pogledamo npr. Eulerovu metodu koja počevši od zadane vrijednosti $y(t_0) = y_0$ generira niz aproksimacija u čvorovima $t_1, t_2, ... , t_{k−3}, t_{k−2}, t_{k−1}, t_k$ onda nam se čini razumnim da pri računanju aproksimacije u $t_{k+1}$ uzmemo u obzir ne samo $t_k, y_k$ nego i dio iz prošlosti, tj. vrijednosti $y_{k−1}, y_{k−2},...$ iz nekoliko prethodnih koraka. Takve metode se zovu višekoračne metode.

### Adams–Bashforthove metode

Prisjetimo se, u jednom koraku metoda koje smo do sada proučavali koristimo relaciju
$$
y(t_{i+1}) = y(t_i) + \int_{t_i}^{t_{i+1}} f(\tau,y(\tau)) d\tau
$$
i u njoj na različite načine numerički računamo integral. Sada pogledamo informaciju nekoliko koraka unazad, recimo u vremenima $t_{k−3}, t_{k−2}, t_{k−1}$, i zajedno sa informacijom u vremenu $t_k$ pokušajmo dobiti bolju procjenu integrala.

Implementirajmo sada Adams–Bashforthovu metodu reda 4.

```python
def AdamsBashforthovaMetoda( f, y0, a, b, h):
    broj_podintervala = int((b - a)/h) 
    
    #dobijanje prvih nekoliko vrijednosti sa rk4
    aa=a
    bb=a+3*h
    [x4,y4]=rk4( f, y0, aa, bb, h)  
    #nastavak algoritma
    x = np.array(x4)
    y = np.array(y4)
    xradni = x[3]
    yradni = np.array([y4[0]])
    
    f0 = np.array(feval(f, x[-1], np.array([y[-1]])))
    f1 = np.array(feval(f, x[-2], np.array([y[-2]])))
    f2 = np.array(feval(f, x[-3], np.array([y[-3]])))
    f3 = np.array(feval(f, x[-4], np.array([y[-4]])))
    
    for i in range(3,broj_podintervala):
        yradni = np.array([y[-1]]) + (h/24)*(55*f0 - 59*f1 +37*f2 - 9*f3)
        xradni += h
        x = np.append(x,xradni)
        y = np.append(y,yradni)            
        fnovi = np.array(feval(f, x[-1], [y[-1]]))
        f3=f2
        f2=f1
        f1=f0
        f0=fnovi
    return [x, y]
```

Za razliku od metode rk4, Adams–Bashforthova metoda u svakom koraku računa samo jednu novu vrijednost funkcije, što je ubrzanje u odnosu na Runge-Kutta metode

```python
#testiranje na primjeru
[ime_metode,vrijeme_izvršavanja,greška,fig]=testProtokol(AdamsBashforthovaMetoda,"Adams-Bashforthova Metoda")
podaci = np.append(podaci,[ime_metode,vrijeme_izvršavanja,greška])
fig
```

Vidimo da smo za ubrzanje algoritma žrtvovali malo na preciznosti.

### Adams-Bashforth-Moultonove Metode

Adams–Moultonove metode polaze od iste ideje kao i Adams–Bashfortove, ali se
ovdje u čvorove inteporlacije uključuje i čvor $t_{i+1}$. Ukupno se gleda isti broj prethodnih koraka, ali s jednim čvorem više možemo konstruirati interpolacijski polinom višeg stupnja. Nadalje, interval nad kojim ćemo integrirati pomoću interpolacijskog polinoma je uključen u interval nad kojim interpoliramo.

```python
def AdamsBashforthMoultonovaMetoda( f, y0, a, b, h):
    broj_podintervala = int((b - a)/h) 
    
    #dobijanje prvih nekoliko vrijednosti sa rk4
    aa=a
    bb=a+3*h
    [x4,y4]=rk4( f, y0, aa, bb, h)  
    #nastavak algoritma
    x = np.array(x4)
    y = np.array(y4)
    xradni = x[3]
    yradni = np.array([y4[0]])
    
    f0 = np.array(feval(f, x[-1], [y[-1]]))
    f1 = np.array(feval(f, x[-2], [y[-2]]))
    f2 = np.array(feval(f, x[-3], [y[-3]]))
    f3 = np.array(feval(f, x[-4], [y[-4]]))
    
    for i in range(3,broj_podintervala):
        xradni += h
        x = np.append(x,xradni)
        yradni = np.array([y[-1]])+ (h/24)*(55*f0 - 59*f1 +37*f2 - 9*f3)
        y = np.append(y,yradni)
        fnovi = np.array(feval(f, x[-1], [y[-1]]))
        yradni = np.array([y[-2]])+ (h/24)*(9*fnovi + 19*f0 - 5*f1 + f2)
        y[-1]=yradni
        
        f3=f2
        f2=f1
        f1=f0
        f0=fnovi
    return [x, y]
```

```python
#testiranje na primjeru
[ime_metode,vrijeme_izvršavanja,greška,fig]=testProtokol(AdamsBashforthMoultonovaMetoda,"Adams-Bashforth-Moultonova Metoda")
podaci = np.append(podaci,[ime_metode,vrijeme_izvršavanja,greška])
fig
```

Smanjili smo grešku sa dosta velikom brzinom algoritma, iz čega nam možemo zaključiti da su Adams-Bashforth-Moultonove metode zapravo jako dobar izbor za rješavanje ODJ

Ostaje nam još usporediti sve metode.

```python
df=pd.DataFrame([podaci[:3],podaci[3:6],podaci[6:9],podaci[9:12],podaci[12:15],podaci[15:]],
              columns=['Naziv', 'Vrijeme', 'Greška'])
df.Vrijeme=pd.to_numeric(df.Vrijeme)
df.Greška=pd.to_numeric(df.Greška)
df['Naziv']=df['Naziv'].str.replace(" Metoda","")
```

## Zaključak

Eulerova metoda je previše neprecizna da bi se uopće mogla dostojno usporediti s ostalim metodama. Modificirana Eulerova je već dala veliki napredak što se tiče točnosti, ali i vremenski je dosta pouzdana. Heunova metoda također daje dobar kompromis preciznosti i vremena izvršavanja, ali imamo i bolje. Ukoliko prioritiziramo preciznost, metoda rk4 je ponajbolje što imamo. Iako malo sporija od Adams-Bashforthove metode, Adams-Bashforth-Moultonova metoda je ipak bolji odabir zbog veće preciznosti i sličnog vremena izvršavanja.

Na kraju treba napomenuti kako na različitim problemima različite metode se pokazuju kao najbolje specijalno za taj problem. S tim umu, ne možemo konkretno zaključiti koja je metoda najbolja, već sa svakim novim problemom trebamo testirati sve metode i odlučuti se za metodu koja nam najviše odgovara.
